```python
import numpy as np

data = {
    "first": ["Ramesh","Rajesh","Mahesh",'Kajoo', np.nan, None, 'NA'],
    "last" : ["Govardhan","singh","pablo",'Singh',np.nan, np.nan, 'Missing'],
    "age": [23,34,22,22,np.nan, None, None],
    "email": ["ramesh@mail.com","rajesh@mail.com","mahesh@mail.com",'kajoor@mail.com', None, np.nan, None],
    "language": [['python','scala','golang'],['scala','golang'],['python','scala','golang'],['scala','golang'], np.nan, None, []],
    "salary":[1000,10029,7298,32793,237927,372973,434342]

}

df = pd.DataFrame(data)
df
```

```python
df.dropna(axis='columns')
```

```python
df.dropna(axis='index')
```

```python
#by default it's set 

df.dropna(axis='index', how='any')
```

```python
df = pd.DataFrame(
    {
        "name": ["Alfred", "Batman", "Catwoman"],
        "toy": [np.nan, "Batmobile", "Bullwhip"],
        "born": [pd.NaT, pd.Timestamp("1940-04-25"), pd.NaT],
    }
)

df
```

```python
#Drop the rows where at least one element is missing.
df.dropna()
```

```python
# Drop the columns where at least one element is missing.
df.dropna(axis="columns")
```

```python
# Drop the rows where all elements are missing.
df.dropna(how="all")
```

```python
df
```

```python
#Keep only the rows with at least 2 non-NA values.
df.dropna(thresh=2)
```

```python
#Define in which columns to look for missing values.

df.dropna(subset=["name", "toy"])
```
