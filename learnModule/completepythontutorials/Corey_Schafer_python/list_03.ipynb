## using functools from compare two value

```python
from functools import cmp_to_key

numbers = [88,9,9909,00,28,00]

def compare_items(x, y):
    # Returns negative if x < y, zero if x == y, positive if x > y
    return x - y 

# This translates a 2-argument function into something Python 3 can use
numbers.sort(key=cmp_to_key(compare_items))
numbers
```

## List Comprehensions

```python
square = []

for num in range(20):
    square.append(num*num)

square
```

```python
square = [x*x for x in range(10)]
square
```

```python
square = map(lambda x: x *x, range(10))

type(square), list(square)
```

```python
list1 = [1, 2, 3]
list2 = [10, 20, 30]

# Multiplies corresponding elements together from two separate lists
products = list(map(lambda x, y: x * y, list1, list2))
print(products) # Output: [10, 40, 90]
```

```python
# Convert a list of numeric strings into actual integers
string_numbers = ["10", "20", "30"]
actual_ints = list(map(int, string_numbers))

print(actual_ints) # Output: [10, 20, 30]
```

```python
# Uppercase an entire list of strings instantly
languages = ["python", "scala", "golang"]
clean_languages = list(map(str.upper, languages))

print(clean_languages) # Output: ['PYTHON', 'SCALA', 'GOLANG']
```

```python
users = [
    {"name": "Ramesh", "age": 23},
    {"name": "Mahesh", "age": 22}
]

# Extract a standalone list of just the names
names_only = list(map(lambda user: user["name"], users))
print(names_only) # Output: ['Ramesh', 'Mahesh']
```
