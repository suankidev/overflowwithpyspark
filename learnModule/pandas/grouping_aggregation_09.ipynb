```python
import pandas as pd
import os
from pathlib import Path
from pathlib import PurePosixPath



BASE_DIR = Path("../").resolve().parent

result_file_path = BASE_DIR / "resources" / "data" / "stack-overflow-survey" / "results.csv"
result_file_schema = BASE_DIR /  "resources" / "data" / "stack-overflow-survey" / "schema.csv"

BASE_DIR,result_file_path,result_file_schema



pd.set_option('display.max_columns',170)
pd.set_option('display.max_rows',170)
pd.set_option('display.max_colwidth',172)
pd.set_option('display.max_seq_items', None)

survey_schema=pd.read_csv(result_file_schema,index_col='qname')
survey_df = pd.read_csv(result_file_path, index_col='ResponseId')
```

```python
data = {
    "first": ["Ramesh","Rajesh","Mahesh",'Kajoo'],
    "last" : ["Govardhan","singh","pablo",'Singh'],
    "age": [23,34,22,22],
    "email": ["ramesh@mail.com","rajesh@mail.com","mahesh@mail.com",'kajoor@mail.com'],
    "language": [['python','scala','golang'],['scala','golang'],['python','scala','golang'],['scala','golang']]

}

df = pd.DataFrame(data)
df
```

```python
exploded_df = df.explode('language')
exploded_df
```

The output 40 means that 40% of all language instances recorded for 22-year-olds are scala (and another 40% are golang).

It is calculating the percentage relative to the total number of skills mentioned, not the total number of unique people!

```python
# 1. Get the total number of unique people for each age from the original dataframe
age_counts = df['age'].value_counts()

# 2. Get the raw counts of languages from the exploded dataframe
lang_counts = exploded_df.groupby('age')['language'].value_counts()

# 3. Divide the language counts by the total number of people in that age bracket
percentage_by_people = (lang_counts / age_counts) * 100
print(percentage_by_people)
```
