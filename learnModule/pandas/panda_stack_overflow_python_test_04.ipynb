## filtering Data 

```python
import pandas as pd

data = {
    "first": ["Ramesh","Rajesh","Mahesh"],
    "last" : ["Govardhan","singh","pablo"],
    "age": [23,34,25],
    "email": ["ramesh@mail.com","rajesh@mail.com","mahesh@mail.com"]

}

df = pd.DataFrame(data)
```

```python
df
```

```python
# want where last name is singh

df['last'] == "singh"
```

```python
name_filter = df['last'] == "singh"

df[name_filter]
```

```python
df.loc[name_filter]
```

```python
df.loc[name_filter,'email']
```

```python
last_name_filter = (df['last'] == 'singh') & (df['first'] == 'Rajesh')
df.loc[last_name_filter]              
```

```python
last_name_filter = (df['last'] == 'singh') | (df['first'] == 'Rajesh')
df.loc[last_name_filter]        
```

```python
df.loc[~last_name_filter]   #opposit of filter
```

### Changing column names

```python
import pandas as pd

data = {
    "first": ["Ramesh","Rajesh","Mahesh"],
    "last" : ["Govardhan","singh","pablo"],
    "age": [23,34,25],
    "email": ["ramesh@mail.com","rajesh@mail.com","mahesh@mail.com"]

}

df = pd.DataFrame(data)
```

```python
df.columns = ['first_name', 'last_name', "age", "email"]
```

```python
df
```

```python
df.columns = [i.upper() for i in list(df.columns)]
```

```python
df
```

```python
df.columns = df.columns.str.replace('_','')
```

```python
df
```

```python
df.FIRSTNAME
```

```python
df.rename(columns={"FIRSTNAME":"first","LASTNAME":"last"}, inplace=True)
```

```python
df
```
