```python
import sys

sys.path.append("C:\\Users\\sujee\\pydev\\pyspark_learn_project\\common_utils")

from sparkUtils import get_spark_session
spark = get_spark_session()

path = "C:\\Users\\sujee\\pydev\\pyspark_learn_project\\resources"
readme_path = path + "\\spark_input\\README.txt"


file_rdd = spark.sparkContext.textFile(readme_path)

file_rdd.collect()


```

```python
#file_rdd_map = file_rdd.map(lambda x: x.split(" ")).flatMap(lambda x: x)
#or
file_rdd_map = file_rdd.flatMap(lambda x: x.split(" "))

file_rdd_map.collect()
```

```python



file_rdd_with_key = file_rdd_map.map(lambda x: (x.lower(), 1))

file_rdd_with_key.take(5)
```

```python
file_rdd_reduce_by_key = file_rdd_with_key.reduceByKey(lambda x,y : x+y)
file_rdd_reduce_by_key.take(5)
```

```python
file_rdd_reduce_by_key.filter(lambda x: (x[0].lower() == 'user') |  (x[0].lower() == '5-star')).collect()
```

```python
for w, k in file_rdd_reduce_by_key.collect():
    print(w,k)
```

```python
flight_path = path + "\\spark_input\\flight_data\\2015-summary.csv"

sc = spark.sparkContext

flight_rdd = sc.textFile(flight_path)

header = flight_rdd.first()

flight_rdd_one = flight_rdd.filter(lambda x: x != header)

flight_rdd_one.take(5)
```

```python
flight_rdd_two = flight_rdd_one.map(lambda x: x.split(","))
flight_rdd_two.map(lambda x:(x[0],x[1],x[2].toInt()))
flight_rdd_two.take(5)
```

```python
flight_rdd_three = flight_rdd_two.groupBy(lambda x: x[0])
flight_rdd_three.take(5)
```

```python
flight_rdd_three.map(lambda x: (x[0], list(x[1])[0][2])).collect()
                     
                   
                     
```
