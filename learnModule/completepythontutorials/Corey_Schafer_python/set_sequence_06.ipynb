```python
a = set('abrakadabra')
a
```

```python
b = set('alacazam')
b
```

```python
a - b ## letters in a but not in b
```

```python
a | b # letter in a and b both
```

```python
a & b   #letter in both a and b
```

```python
## letters in a or b but not both

a ^ b
```

```python
a = {x for x in 'abracadabra' if x not in 'abc'}

type(a)
```
